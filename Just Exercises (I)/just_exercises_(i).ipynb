## Exercise 4 :  A function for Fibonacci

```python
#method 1 recursion

def fibonacci(n):
    if n==1:
        return [0]
    if n==2:
        return [0,1]
    prev = fibonacci(n-1)
    prev.append(prev[-2] + prev[-1])
    return prev

print(fibonacci(15))

#method 2 iteration

def fib2(n):
    l = []
    a,b = 0,1
    for i in range(n):
        l.append(a)
        a,b = b, a+b
    return l
print(fib2(15))
```

## Exercise 5 :  max_info() without NumPy

```python
def max_info(numbers):
    my_max = numbers[0]
    my_max_index = 0
    for i in range(len(numbers)):
        if numbers[i] > my_max:
            my_max = numbers[i]
            my_max_index = i
    return(my_max, my_max_index)

max_info([40, 27, 83, 44, 74, 51, 76, 77, 10, 49])
            
```
