## Exercise 1 :  F-strings to the rescue

```python
for i in range(1,11):
    print(f"{i:05d}")
```

```python
for i in range(1,11):
    string = f"{i:03d}"
    a = "+"
    string += f"{a*i:>10s}"
    print(string)
```

```python
for i in range(1,11):
    string = f"{i:03d} "
    a, b = "-", "+"
    string += f"{a*i+b*(10-i):<10s}"
    print(string)
```
