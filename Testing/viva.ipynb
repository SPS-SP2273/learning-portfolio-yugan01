```python
import glob
import os
import matplotlib.pyplot as plt
import numpy as np

os.getcwd()
#os.chdir('golgi-movie_robert_frames_gray')

img_data = plt.imread("golgi-movie_robert_frame_15_gray.jpg")
plt.imshow(img_data, cmap='jet')
plt.colorbar()
plt.show()
```

```python
type_of_data = str(type(img_data))[8:-2] + ' uint8'
flattened = img_data.flatten()
min_value = min(flattened) #pixel value of the darkest pixel
max_value = max(flattened) # pixel value of the lightest pixel
sum_value = sum(flattened) # sum of all the pixel values in the 512x512 area
print(type_of_data) #np.ndarray uint8
print(img_data.shape) # (512, 512), 512x512 pixels
print(min_value, max_value, sum_value) #0 222 1123362
plt.hist(flattened, bins = 100)
plt.show() #right skewed distribution with the highest porportion of dark/black pixels
```

```python
dict_of_type = dict()
dict_of_shape = dict()
dict_of_min = dict()
dict_of_max = dict()
dict_of_sum = dict()
list_of_images = glob.glob('*.jpg')
list_of_images[0].rsplit('_gray.jpg')[0][-2:]
for image in list_of_images:
    img_data = plt.imread(image)
    key = 'image ' + image.rsplit('_gray.jpg')[0][-2:]
    type_of_data = str(type(img_data))[8:-2]
    dict_of_type[key] = f'{type_of_data} {img_data.dtype}'
    dict_of_shape[key] = img_data.shape
    flattened = img_data.flatten()
    dict_of_min[key] = min(flattened) # the pixel value of the darkest pixel 
    dict_of_max[key] = max(flattened) # the pixel value of the lighest pixel
    dict_of_sum[key] = sum(flattened) # sum of all pixel values in the image
```
